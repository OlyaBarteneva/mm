```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def dydt(y, t):
    k = 0.42
    return k * y
```

```python
h = 0.1
t = np.arange(0, 10 + h, h)
print(t)
```

```python
def euler(f, y0, t):
    y = [0] * len(t)
    y[0] = y0
    for i in range(1, len(t)):
        y[i] = y[i-1] + h * f(y[i-1],t[i-1])
    return y
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
solve_euler = euler(dydt, 1, t)
print(len(solve_euler), len(solve))
```

```python
r = [0] * len(t)
for i in range(0, len(t)):
    r[i] = solve[i] - solve_euler[i]
print(r)
```

```python
x = np.arange(1, 102, 1)
%matplotlib inline
plt.plot(x, r, label = r"Разности")
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y'= -ky$")
plt.plot(t, solve_euler, label = r"Метод Эйлера")
plt.grid(True)
plt.legend()
plt.show()
```
