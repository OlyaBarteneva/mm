**Простейшая модель роста Мальтуса (для роста населения Земли)**

$y'=-ky$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def dydt(y, t):
    k = 0.42
    return -k * y
```

```python
t = np.linspace(0, 10, 100)
print(t)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y'= -ky$")
plt.grid(True)
plt.legend()
plt.show()
```

**Мягкая модель роста с зависящим от населения коэффициентом размножения** 

$y'=k(y)y$

```python
def k(y):
    a = 0.42
    b = 0.31
    return a - b * y
def dydt2(y, t):
    return k(y) * y
```

```python
solve = odeint(dydt2, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y'= k(y)y$")
plt.grid(True)
plt.legend()
plt.show()
```

**Модель изменения популяции рыб при рыболовстве с интенсивностью $c=- 0.42$**

$y'=y-y^2-c$

```python
def dydt3(y, t):
    c = - 0.42
    return y - y ** 2 - c
```

```python
solve = odeint(dydt3, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y'= y-y^2-c$")
plt.grid(True)
plt.legend()
plt.show()
```

**Мягкая модель изменения популяции рыб**

$y'=y-g(y)y-c$

```python
def g(y):
    a = 0.6
    b = 0.0009
    return a - b * y
def dydt4(y, t):
    c = 0.7
    return y - g(y) * y - c
```

```python
solve = odeint(dydt4, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y'= y-g(y)y-c$")
plt.grid(True)
plt.legend()
plt.show()
```

**Модель изменения популяции рыб при дифференциальной квоте вылова**

$y'=y-y^2-ky$

```python
def dydt5(y, t):
    k = 0.7
    return y - y ** 2 - k * y
```

```python
solve = odeint(dydt5, y0=1, t=t)
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$y'= y-y^2-ky$", color = "red")
plt.grid(True)
plt.legend()
plt.show()
```
