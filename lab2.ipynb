```python
def dydt(y, t):
    k = 0.02
    return k * y

def k(y):
    a = 40000000
    b = 300
    return a - b * y
def dydt2(y, t):
    return k(y) * y
```

```python
solve = odeint(dydt, y0=4000, t=years)
solve = solve.flatten()

solve2 = odeint(dydt2, y0=4000, t=years)
solve2 = solve2.flatten()
```
